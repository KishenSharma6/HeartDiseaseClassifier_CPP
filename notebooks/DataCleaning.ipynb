### Data Cleaning

```python
#Copy raw_df
clean_df = raw_df.copy()
```

```python
#Update data types from int64 to object
discrete = ["sex", "cp", "fbs", "restecg", "exang", "slope",
            "ca", "thal", "target", ]

for col in discrete:
    clean_df[col] = clean_df[col].astype("object")
```

```python
#Encode Variables
for col in discrete:
    if clean_df[col].nunique() <=2:
        pass
    else:
        one_hot = pd.get_dummies(clean_df[col], prefix= [col])
        clean_df= clean_df.drop(col, axis= 1)
        clean_df= pd.concat([clean_df,one_hot], axis = 1)

clean_df.head();
        
        
        
```

```python
#Shuffle data
from sklearn.utils import shuffle
clean_df = shuffle(clean_df, random_state= 24, )
clean_df.head()
```
